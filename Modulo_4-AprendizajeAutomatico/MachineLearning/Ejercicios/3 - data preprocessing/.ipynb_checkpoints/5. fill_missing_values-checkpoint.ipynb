# Rellenar Missing Values



La mayora de los modelos de aprendizaje automtico no aceptan valores no informados en los datos de entrada. 

En las variables categricas, los valores que faltan pueden transformarse en una nueva categora "MISSING", resolviendo as el problema. Tambin pueden sustituirse por la moda.

En el caso de las variables numricas, estos valores perdidos pueden rellenarse utilizando distintos enfoques, como rellenar con la media o mediante diversos estimadores.



## Load Data

```python
dat = pd.read_csv("../datasets/i2.csv", sep = ";")
dat
```

## Detectar Missing Values

Contemos el nmero de valores no informados de cada variable del conjunto de datos.

```python
dat.apply(lambda x: 100*np.sum(x.isna())/len(x))
```

origin_airport y ant tienen un bajo porcentaje de valores no informados. Deberamos intentar rellenarlos.

## Approach 0: Eliminar filas missing values.

El enfoque ms bsico sera simplemente eliminar cualquier fila con valores perdidos. Sin embargo, aunque este enfoque se recomienda a veces en cursos y tutoriales, no es aconsejable en la vida real. Hacer esto probablemente te dejar casi sin datos ;p.

```python
dat_new = dat.copy()
dat_new.dropna(axis = 0, how = 'any', inplace = True)
dat_new
```

## Variables Categricas

```python
dat.dtypes
```

```python
categorical_variables = dat.columns.values[dat.dtypes == 'object'].tolist()
numerical_variables = dat.columns.values[dat.dtypes == 'float64'].tolist()
```

```python
categorical_variables
```

```python
numerical_variables
```

### Approach 1: Reemplazar con la moda.

El siguiente enfoque en relacin con las variables categricas ser sustituir los valores que faltan por la categora ms frecuente o la moda.

```python
modes = dat[categorical_variables].apply(lambda x: stats.mode(x)[0][0]).to_dict()
modes
```

Utilicemos los valores de la moda para sustituir los NA.

```python
dat
```

```python
dat_new = dat.fillna(value = modes, axis = 0)
dat_new
```

```python
dat_new.apply(lambda x: 100*np.sum(x.isna())/len(x))
```

### Approach 2: Transformar a una nueva categora.

Los valores que faltan en una variable categrica pueden aportar informacin, especialmente si no son aleatorios. Por lo tanto, puede ser interesante considerar los valores no informados como una nueva categora en lugar de sustituirlos por la moda o algn otro valor.

```python
dat_new = dat.copy()
dat_new[categorical_variables] = dat[categorical_variables].fillna(value = 'UNKNOWN', axis = 0)
dat_new
```

```python
dat_new.apply(lambda x: 100*np.sum(x.isna())/len(x))
```

## Variables Numricas

### Approach 1: Reemplazar con la media.

El siguiente enfoque ms bsico en relacin con las variables numricas ser sustituir los valores que faltan por el valor medio o la mediana.

```python
means = dat[numerical_variables].apply(lambda x: np.mean(x)).to_dict()
means
```

Utilicemos estos valores medios para sustituir los NA.

```python
dat_new = dat.fillna(value = means, axis = 0)
dat_new
```

```python
dat_new.apply(lambda x: 100*np.sum(x.isna())/len(x))
```

### Approach 2: Usar un algoritmo..

En cuanto al relleno de valores no informados de variables numricas, existen mtodos mucho ms complejos para rellenarlos, algunos de ellos incluso utilizan modelos de aprendizaje automtico para intentar predecir estos missing values.

Este es el caso de la MICE: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3074241/. 

```python
from fancyimpute import IterativeImputer as MICE # pip install fancyimpute
?MICE
```

```python
dat_new = dat.copy()
dat_new[numerical_variables] = MICE().fit_transform(dat[numerical_variables])
dat_new
```

```python
dat_new.apply(lambda x: 100*np.sum(x.isna())/len(x))
```
