## Visualizaciones

Vamos a realizar ahora un ejemplo de cada una de las visualizaciones vistas en teoría.

### Variables numéricas

#### Scatter plot

Primero realizamos un scatter plot que muestre *Sepal_Width* con respecto a *Sepal_Length*. Para ello utilizaremos la librería *matplotlib*

```python
plt.scatter(x = dat.Sepal_Length, y = dat.Petal_Length)
plt.show()
```

Existen múltiples parámetros para personalizar nuestros gráficos en *matplotlib*. Por ejemplo, podemos modificar el color de los puntos.

```python
plt.scatter(x = dat.Sepal_Length, y = dat.Petal_Length, c = 'red')
plt.show()
```

O hacerlos depender de otra variable, como el tipo de flor.

```python
colors = {'setosa':'red', 'virginica':'green', 'versicolor':'blue'}
plt.scatter(x = dat.Sepal_Length, y = dat.Petal_Length, c = dat.Species.map(colors))
plt.show()
```

#### Line plot

```python
# Draw me a line plot per flower species coloured independently

plt.plot(dat.Petal_Length)
plt.show()
```

Cambiamos el tipo de línea.

```python
plt.plot(dat.Petal_Length, linestyle='dashed')
plt.show()
```

Pandas permite realizar visualizaciones simultáneas de todas las variables.

```python
dat.plot.line()
```

#### Histograma

```python
plt.hist(dat.Petal_Length)
plt.show()
```

Recordad que podemos comprobar los parámetros que se pueden personalizar de cualquier visualización llamando al comando help (?) de Jupyter.

```python
plt.hist?
```

```python
plt.hist(dat.Petal_Length, bins = 4)
plt.show()
```

#### Boxplot

```python
plt.boxplot(dat.Petal_Length)
plt.show()
```

Lo mostramos ahora en horizontal.

```python
plt.boxplot(dat.Petal_Length, vert = False)
plt.show()
```

#### Función de densidad y función de distribución

Para estas gráficas utilizaremos la librería *seaborn*.

```python
sns.distplot(dat.Petal_Length)
```

Podemos eliminar el histograma de fondo.

```python
sns.distplot(dat.Petal_Length, hist = False)
```

### Variables categóricas

#### Barplot

Para visualizar un barplot primero debemos calcular la tabla de frecuencias de la variable.

```python
tf = dat.Species.value_counts()
tf
```

```python
tf.index
```

```python
tf.values
```

```python
plt.bar(tf.index, tf.values)
```

#### Pie Chart

```python
plt.pie(tf.values)
plt.show() 
```

Podemos añadir los labels.

```python
plt.pie(tf.values, labels = tf.index)
plt.show() 
```
